```
import pandas as pd 
import seaborn as sns

iris = sns.load_dataset('iris')

iris.head()
```

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```
# scatterplot 
sns.scatterplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

http://seaborn.pydata.org/generated/seaborn.jointplot.html

```
# joint plot
sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

https://seaborn.pydata.org/generated/seaborn.boxplot.html

```
# box plot 
sns.boxplot(x = 'species', y = 'sepal_length', data = iris)
```

https://seaborn.pydata.org/generated/seaborn.barplot.html

```
#Barplot
sns.barplot(x = 'species', y = 'sepal_length', data = iris)
```

```
sns.set_theme()
sns.jointplot(data=iris, x = 'sepal_width', y = 'petal_width', hue = 'species', kind = 'kde', palette='inferno')
```

```
sns.set_theme()
sns.kdeplot(data=iris, x = 'sepal_width', y = 'petal_width', hue = 'species', palette='dark')
```

```
from numpy import mean
sns.barplot(data=iris, x='species', y='sepal_length', estimator=mean, order=['virginica', 'versicolor','setosa'])
```
